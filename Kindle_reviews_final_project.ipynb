```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Downloads/kindle_review.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['helpful', 'reviewTime', 'reviewerName','summary','unixReviewTime'],axis=1)
df
```

```python
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
stop_words=set(stopwords.words('english'))
```

```python
def sentiment_calc(reviewText):
    try:
        return TextBlob(str(reviewText)).sentiment.polarity
    except:
        return None
df['sentiment'] = df['reviewText'].apply(sentiment_calc)
df.head(50)
```

```python
df['score']=df['overall']+df['sentiment']
```

```python
def classify_score(score):
    if score>=5 and score<= -0.75:
        return 1
    elif score>-0.75 and score<0:
        return 2
    elif score>=0 and score<2:
        return 3
    elif score>=2 and score<3.5:
        return 4
    else:
        return 5
df['new_score']=df['score'].apply(classify_score)
df.head()
```

```python
df_pivot=df.pivot_table(index='asin', columns='reviewerID', values='overall').fillna(0)
df_pivot
```

```python
from scipy.sparse import csr_matrix
df_pivot_matrix=csr_matrix(df_pivot.values)
print(df_pivot_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine')
model_knn.fit(df_pivot_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix=cosine_similarity(df_pivot)
similarity_matrix
```

```python
prod_id=list(df_pivot.index)
print(prod_id)
```

```python
product_id=input("Enter product id according to dataset: ")
product_index=prod_id.index(product_id)

print(product_index)
```

```python
similarity, indices=model_knn.kneighbors(df_pivot.iloc[product_index, :].values.reshape(1,-1),n_neighbors=8)
print(similarity)
print(indices)
```

```python
data_dict={}
for i in range(0, len(similarity.flatten())):
    if i==0:
        print("recommendations for {0}: \n".format(df_pivot.index[product_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])]=float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, is similarity distance=with {similarity.flatten()[i]:.20f}:')
print(data_dict)
```
